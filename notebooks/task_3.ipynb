```python
import pandas as pd
```

```python
df=pd.read_csv('Reviews.csv')
df.head()
```

```python
df['Score'].value_counts()
```

```python
df=df.drop(["Id","ProductId","UserId","ProfileName","HelpfulnessNumerator","HelpfulnessDenominator","Time","Summary"],axis=1)
df.head()
```

```python
df.duplicated().sum()
```

```python
# Dropping conflicting reviews
conflict_texts = df.groupby("Text")["Score"].nunique()
conflict_texts = conflict_texts[conflict_texts > 1].index
# conflict_texts

```

```python
df1 = df[~df["Text"].isin(conflict_texts)]
```

```python
# Dropping exact Duplicates
df1 = df1.drop_duplicates(subset=["Text", "Score"])
df1
```
