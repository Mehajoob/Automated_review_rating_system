```python
# #Creating imbalanced data
# N = 25000 

# ratios = {
#     1: 0.10,   
#     2: 0.15,  
#     3: 0.25,  
#     4: 0.30,  
#     5: 0.20   
# }

# ratings = []

# for rating, pct in ratios.items():
#     count = int(N * pct)
#     ratings.extend([rating] * count)

# df_imbalanced = pd.DataFrame({'Score': ratings})

# df_imbalanced = df_imbalanced.sample(frac=1).reset_index(drop=True)

# # df_imbalanced.head()
# df_imbalanced['Score'].value_counts()  

# Creating imbalanced data
N = 25000

# Ratios for each score
ratios = {
    1: 0.10,
    2: 0.15,
    3: 0.25,
    4: 0.30,
    5: 0.20
}

# Empty list to store subsets
dfs = []

for rating, pct in ratios.items():
    count = int(N * pct)    # samples to take for this rating
    subset = df[df['Score'] == rating].sample(n=count, random_state=42)
    dfs.append(subset)

# Combine all subsets
df_imbalanced = pd.concat(dfs).sample(frac=1, random_state=42).reset_index(drop=True)

df_imbalanced.shape
df_imbalanced['Score'].value_counts()


```

```python
# visualization after imbalancing

#  Countplot with Annotation (Improved Bar Plot)

sns.countplot(
    x='Score', 
    data=df_imbalanced,
    hue='Score',
    palette='viridis',
    legend=False
)

plt.title("Imbalanced Rating Distribution with Counts")
plt.show()
```

```python
# Pie Chart of Imbalanced Ratings
df_imbalanced['Score'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Imbalanced Rating Distribution (Pie Chart)")
plt.ylabel("")
plt.show()
```

```python
sns.kdeplot(df_imbalanced['Score'])
plt.title("Rating Density Plot")
plt.show()
```

```python
counts = df_imbalanced['Score'].value_counts().sort_index()
cum_percent = counts.cumsum() / counts.sum() * 100
cum_percent.plot(marker='o')

plt.title("Cumulative Rating Distribution")
plt.ylabel("Cumulative Percentage (%)")
plt.xlabel("Rating")
plt.grid(True)
plt.show()
```
