```python
# Rating distribution(bar chart)
ratings = df['Score'].value_counts().sort_index()

plt.bar(ratings.index, ratings.values)
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
df['review_length'] = df['Text'].str.len()

sns.violinplot(data=df, x="Score", y="review_length", inner="quartile")
plt.title("Violin Plot of Review Length per Rating")
plt.show()
```

```python
df['Score'].value_counts()
```

```python
#splitting into equal counts(5000)
df_balanced = df.groupby('Score').sample(n=5000, random_state=42)
df_balanced
```

```python
df_balanced['Score'].value_counts()
```

```python
# Rating distribution(bar chart)
ratings = df_balanced['Score'].value_counts().sort_index()

plt.bar(ratings.index, ratings.values)
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Score', y='review_length', data=df_balanced)
plt.title("Review Length vs Rating")
plt.show()
```

```python
plt.hist(df_balanced['review_length'], bins=50)
plt.title("Review Length Distribution")
plt.show()
```
